```python
import pandas as pd

# CSVファイルを読み込み、DataFrameに格納
df_1 = pd.read_csv(r'C:\Users\iwai-ryuya\Downloads\あつまる・ぱちまる予定.csv')
```

```python
display(df_1)
```

```python
sum_medal = df_1["差"].sum() *-1
print(sum_medal)
sum_medal_text = f'総差枚は{sum_medal}枚でした。'
print(sum_medal_text)
```

```python
# 平均ゲーム数,

# 総台数,総アウトの変数を作成する。
#　総アウト変数/総台数変数=平均アウト変数
# 平均アウト変数/3

#総台数変数の作成
sum_machine = len(df_1)
print(sum_machine)

#総アウト変数の作成
sum_out = df_1["アウト"].sum()
print(sum_out)

#平均アウト変数の作成
ave_out = sum_out/sum_machine
print(ave_out)


ave_game = ave_out/3
print(ave_game)
ave_game = int(ave_out/3)
ave_game_text = f'平均ゲーム数は{ave_game}Gでした。'
print(ave_game_text)
```

```python
##総平均差枚

ave_medal = sum_medal/sum_machine
print(ave_medal)
ave_medal = int(sum_medal/sum_machine)
ave_medal_text = f'1台あたりの平均枚数は{ave_medal}枚でした。'
print(ave_medal_text)
```

```python
# 機種別平均差枚

# 機種名がどれだけあるかを調べる

machine_ave_medal = df_1.groupby('機種名')['差'].mean()*-1
display(machine_ave_medal)
```
